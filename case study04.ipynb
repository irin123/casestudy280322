```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# read the data set


```python
#1. Read the dataset to the python environment.
df = pd.read_csv('cars_data.csv')
```

```python
df.head(10)
```

# null values

```python
#2.check for the null values present in the dataset
df.isnull().sum()
```

# Bar Graph

```python
#3.Plot a bar graph of male vs female buyers participated in the sales.
plt.figure(figsize=(8,8))
plt.bar(df['Buyer Gender'],height=1,width=0.5)
plt.title('MALE BUYERS Vs FEMALE BUYERS',fontsize=20)
plt.yticks()

```

# Find the top 5 cars based on their sales price

```python
sorted_data = df.sort_values(by=['Sale Price'], ascending=False)
sorted_data.head(5)
```

# Find the least 5 cars based on their Resell price.


```python
sorted_data = df.sort_values(by=['Resell Price'],ascending=True)
sorted_data.head(5)
```
